#### Correlation between Subscribers and Likes, Subscribers and Comments

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming youtubers_df is your DataFrame

# Calculate the correlation between 'Suscribers' and 'Likes'
corr_likes = youtubers_df['Suscribers'].corr(youtubers_df['Likes'])
print(f"Correlation between number of subscribers and number of likes: {corr_likes}")

# Calculate the correlation between 'Suscribers' and 'Comments'
corr_comments = youtubers_df['Suscribers'].corr(youtubers_df['Comments'])
print(f"Correlation between number of subscribers and number of comments: {corr_comments}")

# Select numeric columns
numeric_cols = youtubers_df.select_dtypes(include=[np.number])

# Calculate the correlation matrix
corr_matrix = numeric_cols.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Content Categories

```python


# Assuming youtubers_df is your DataFrame
# If not, replace it with your actual DataFrame

# Count the number of streamers per category
category_counts = youtubers_df['Categories'].value_counts()

# Visualize the distribution of content categories
plt.figure(figsize=(15, 8))
sns.barplot(x=category_counts.values, y=category_counts.index, palette='viridis')
plt.title('Distribution of Content Categories')
plt.xlabel('Number of Streamers')
plt.ylabel('Categories')
plt.show()

# Identify categories with exceptional performance metrics
exceptional_categories = youtubers_df.groupby('Categories')[['Suscribers', 'Visits', 'Likes', 'Comments']].mean()
exceptional_categories['PerformanceMetric'] = exceptional_categories.mean(axis=1)
exceptional_categories = exceptional_categories.sort_values(by='PerformanceMetric', ascending=False)

# Visualize exceptional categories
plt.figure(figsize=(12, 6))
sns.barplot(x=exceptional_categories['PerformanceMetric'], y=exceptional_categories.index, palette='magma')
plt.title('Exceptional Performance Metrics by Category')
plt.xlabel('Average Performance Metric')
plt.ylabel('Categories')
plt.show()
```

#### Brands and Collaborations

```python

# Filter out rows with missing values in 'Links' column
youtubers_with_links = youtubers_df.dropna(subset=['Links'])

# Calculate a performance metric for each streamer
youtubers_with_links['PerformanceMetric'] = youtubers_with_links[['Suscribers', 'Visits', 'Likes', 'Comments']].mean(axis=1)

# Visualize the relationship between performance metrics and brand collaborations
plt.figure(figsize=(12, 6))
sns.scatterplot(x='PerformanceMetric', y='Links', data=youtubers_with_links, color='blue', alpha=0.5)
plt.title('Relationship between Performance Metrics and Brand Collaborations')
plt.xlabel('Average Performance Metric')
plt.ylabel('Brand Collaborations (Links)')
plt.show()
```

#### Benchmarking

```python
# Assuming youtubers_df is your DataFrame
# If not, replace it with your actual DataFrame

# Calculate average performance metric
avg_performance = youtubers_df[['Suscribers', 'Visits', 'Likes', 'Comments']].mean(axis=1)

# Add average performance metric to the DataFrame
youtubers_df['AveragePerformance'] = avg_performance

# Identify top-performing content creators
top_performers = youtubers_df[youtubers_df['AveragePerformance'] > avg_performance.mean()]

# Create a plot
plt.figure(figsize=(12, 8))
sns.barplot(x='AveragePerformance', y='Username', data=top_performers, palette='viridis')
plt.title('Top-Performing Content Creators')
plt.xlabel('Average Performance Metric')
plt.ylabel('Username')
plt.show()

```

```python

ranked_data = top_performers.sort_values(by="AveragePerformance", ascending=False)

# Reset the index to reflect the new ranking
ranked_data = ranked_data.reset_index(drop=True)
ranked_data
```
